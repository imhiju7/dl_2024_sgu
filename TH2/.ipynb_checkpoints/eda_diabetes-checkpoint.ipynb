###**3. Data Descriptive**

```python
df.head()
```

**Data summary**
- **`Pregnancies`** (Số lần mang thai): Số lần mang thai của bệnh nhân nữ trong dữ liệu. Đây có thể là một yếu tố quan trọng liên quan đến nguy cơ mắc bệnh tiểu đường, đặc biệt với phụ nữ có số lần mang thai nhiều.
- **`Glucose`** (Nồng độ đường trong máu): Mức glucose đo được sau 2 giờ xét nghiệm dung nạp glucose. Giá trị này có thể trực tiếp liên quan đến việc chẩn đoán bệnh tiểu đường.
- **`BloodPressure`** (Huyết áp tâm trương): Huyết áp tâm trương (mm Hg). Các mức huyết áp bất thường có thể ảnh hưởng đến tình trạng sức khỏe, bao gồm cả nguy cơ mắc bệnh tiểu đường.
- **`SkinThickness`** (Độ dày da): Độ dày nếp gấp da (mm). Đây là một yếu tố thể hiện mức độ mỡ dưới da, có liên quan đến tình trạng béo phì và nguy cơ mắc tiểu đường.
- **`Insulin`** (Mức insulin trong huyết thanh): Mức insulin trong huyết thanh sau 2 giờ (mu U/ml). Insulin có thể giúp xác định khả năng điều hòa đường huyết của cơ thể.
- **`BMI`** (Chỉ số khối cơ thể): Chỉ số khối cơ thể, được tính bằng trọng lượng (kg) chia cho bình phương chiều cao (m). `BMI` cao thường liên quan đến nguy cơ mắc tiểu đường loại 2.
- **`DiabetesPedigreeFunction`** (Chức năng phả hệ tiểu đường): Giá trị này thể hiện nguy cơ mắc bệnh tiểu đường dựa trên di truyền và lịch sử gia đình.
- **`Age`** (Tuổi): Tuổi của bệnh nhân, thường là một yếu tố quan trọng trong việc xác định nguy cơ mắc bệnh tiểu đường.
- **`Outcome`** (Kết quả): Cột nhị phân với giá trị 0 (không mắc bệnh) và 1 (mắc bệnh tiểu đường).

###**4. Exploratory Data Analysis**

```python
# Count the occurrences of each outcome (0 = non-diabetic, 1 = diabetic)
diabetes_counts = df['Outcome'].value_counts()

# Define labels for the pie chart
labels = ['Non-Diabetic (0)', 'Diabetic (1)']

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(diabetes_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#ff9999', '#66b3ff'])
plt.title('Diabetes Prevalence Rate')  # Title of the pie chart
plt.axis('equal')  # Ensure that pie chart is a circle
plt.show()  # Display the pie chart
```

- 65.1% không bị mắc bệnh tiểu đường và 34.9% mắc bệnh tiểu đường




```python
from prettytable import PrettyTable

descriptive_stats = df.groupby('Outcome').agg(['mean', 'std'])
# Sử dụng Styler để định dạng bảng
styled_table = descriptive_stats.style.set_caption("Mean vs Std")\
    .format(na_rep='NaN', precision=2)\
    .background_gradient(cmap='viridis')

# Hiển thị trong Jupyter Notebook
styled_table
```

```python
# Define a list of features to plot, excluding 'Outcome'
features = df.columns[df.columns != 'Outcome'].tolist()

# Create a figure to plot each feature
plt.figure(figsize=(16, 10))
for i, feature in enumerate(features, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df[feature], kde=True, bins=10, color='red')
    plt.title(f'Distribution of {feature}')

plt.tight_layout()
plt.show()
```


**1. Pregnancies**:

**3.BloodPressure**:

   - Một vài giá trị bằng 0 có thể là dữ liệu bị thiếu hoặc bất thường.

**4.SkinThickness**:
   - Phân phối lệch phải với nhiều giá trị bằng 0, cho thấy có khả năng dữ liệu bị thiếu.
   - Đa số giá trị nằm trong khoảng từ 20–40 mm.

**5.Insulin**:
   - Phân phối lệch mạnh về phía phải với rất nhiều giá trị bằng 0.
   - Các giá trị insulin tăng đột biến từ 0 đến khoảng 600 µU/mL.

**6.BMI**:

**7.DiabetesPedigreeFunction**:

**8.Age**:
